```python
import numpy

import matplotlib
import matplotlib.pyplot as plt

import librosa
from IPython.display import Audio
import IPython
```

```python
sr = 44100

song_name = 'ww_the_deep.wav'
x_full, sr = librosa.load(f'./datasets/{song_name}', sr=sr, mono=True)
x = x_full[1000000 : 1000000 + sr * 5]

Audio(data=x, rate=sr)
```

```python
from neurowhore.utils import mulaw, mulraw_inverse, simple_quantizer

y = mulaw(x)
x_rec = mulraw_inverse(y)
```

```python
from visual.plotters import plot_waveforms
plot_waveforms([x, y, x_rec])
```

```python
q_ch = 2**8
x_q = simple_quantizer(x, quantization_channels=q_ch)
y_q = simple_quantizer(y, quantization_channels=q_ch)
x_q_rec = mulraw_inverse(y_q)

plot_waveforms([x_q, y_q, x_q_rec])
```

```python
IPython.display.display(
    IPython.display.Audio(x, rate=sr)
)
IPython.display.display(
    IPython.display.Audio(x_q, rate=sr)
)
IPython.display.display(
    IPython.display.Audio(x_q_rec, rate=sr)
)
```
